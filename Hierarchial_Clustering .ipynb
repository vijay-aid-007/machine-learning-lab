```python
# IMPORTNIG NECESSARY PACKAGES 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.cluster.hierarchy import dendrogram , linkage , fcluster
```

```python
# 1. GENERATE OR LOAD THE DATASET
x = np.array([
    [1,2],
    [2,3],
    [3,3],
    [8,7],
    [8,8],
    [7,7],
    [1,0],
    [0,1],
    [9,8],
    [8,9]

])
x
```

```python
# 2 PERFROM HIERARCHIAL CLUSTERING 
"""
LINKAGE FUNCTION BUILDS THE HIERARCHIAL CLUSTERS 
METHOD = "WARD" -> MINIMIZES THE VARIANCE OF CLUSTERS BEING MERGED 
METRIC = 'EUCLIDEAN' -> DISTANCE METRIC (CAN ALSO USE OTHERS DISTANCE METRICS SUCH AS , MANHATTAN , COSINE ETC.....)
"""

Z = linkage(x, method='ward', metric='euclidean')
Z
```

```python
"""
3. PLOT THE DENDROGRAM 
"""
plt.figure(figsize=(8,5))
plt.title('"Hierarchial_Clustering')
plt.xlabel('Data Point Index')
plt.ylabel('Euclidean Distance')

#DENDROGRAM DRAWS THE TREE 
dendrogram(Z , leaf_rotation=350, leaf_font_size=12)
plt.show()
```

```python
"""
4. DECIDE THE NUMBER OF CLUSTERS AND ASSIGN CLUSTER LABELS 

"""

clusters = fcluster(Z , t = 3 , criterion='maxclust')
print(f"Cluster labels for each data point : {clusters}")
```

```python
""" 
5. (Optional) Visualize clusters in 2D
"""
plt.figure(figsize=(6,5))

"""
the parameters s controls the size of the  markers in the scatter plot 
cmap = 'rainbow' , assigns the colors to the labels  which are drawn from the rainbow colors
"""
plt.scatter(x[:,0] , x[:, 1] , c = clusters , cmap='rainbow' , s = 100)
plt.title('Data Points Grouped By Heirarchial Clustering')
plt.xlabel('Features 1')
plt.ylabel('Features 2')
plt.show()
```
